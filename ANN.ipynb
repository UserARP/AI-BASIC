```python
import tensorflow as tf
```

```python
import numpy as np
import datetime
from tensorflow.keras.datasets import fashion_mnist
```

```python
tf.__version__
```

```python
(X_train,y_train),(X_test,y_test)= fashion_mnist.load_data() 
#เก็บด้าต้าไว้ในตัวแปรสี่ตัวนี้
```

```python
X_train = X_train / 255
X_test = X_test / 255
#normalize หรือทำการแปลงค่าให้ข้อมูลอยู่ระหว่าง 0-1
```

```python
#่ต่อไปทำการ reshape ค่าของด้าต้าให้เป็นค่ามิติเดียว
X_train = X_train.reshape(-1,28*28)
```

```python
X_train.shape
```

```python
X_test = X_test.reshape(-1,28*28)
```

```python
## สร้าง model##
```

```python
#ประกาศ โมเดล
model = tf.keras.models.Sequential()
```

```python
#เพิ่ม layer แรกของโมเดล(Dense layer)
model.add(tf.keras.layers.Dense(units=128,activation='relu',input_shape=(784, )))
```

```python
#เพิ่ม layer ที่สอง(Dropout layer)
model.add(tf.keras.layers.Dropout(0.2))
```

```python
#เพิ่ม layer สุดท้ายlayerที่สาม (output layer)
model.add(tf.keras.layers.Dense(units=10,activation='softmax'))
```

```python
#compile model (optimizer:Adam)
model.compile(optimizer='adam',loss = 'sparse_categorical_crossentropy',metrics=['sparse_categorical_accuracy'])
```

```python
model.summary()
```

```python
#training โมเดล
model.fit(X_train,y_train,epochs=200)
```

```python
#ประเมินโมเดลที่ทำออกมาโดยนำข้อมูลส่วนของx_testมาประเมิน
 #กำหนดตัวแปรสำหรับแสดง
test_loss,test_acc = model.evaluate(X_test,y_test)
```

```python
print("Test acc : {}".format(test_acc))
```
